## Modelo de propagación de una epidemia (Susceptibles-Infectados / SI model)

Un sistema de dos compartimientos poblaciones (susceptibles de infección $S(t)$ e infectados $I(t)$), para el que la condición inicial es $I(t=0) = 1$, tiene una solución dada por las siguientes funciones:

$$ I(t) = \frac{N}{1-(1-N)e^{-N\beta t}} $$

$$ S(t) = N - I(t) $$

```
import numpy as np 
import matplotlib.pyplot as plt

# poblacion total fija
n = 100.0
# parametros modificables 
i0 = 1.0
beta = 0.003
def i(i0, beta, t): return n*i0/(i0-(i0-n)*np.exp(-n*beta*t))
def s(*args): return n - i(*args)
```

```
# @title Tablas de datos
t_array = np.arange(0,50,1)
i_array = i(i0, beta, t_array)
s_array = s(i0, beta, t_array)
print(t_array)
print(i_array)
```

```
# @title visualizacion solución
plt.plot(t_array, i_array, label = 'infectados')
plt.plot(t_array, s_array, label = 'susceptibles')
plt.grid(True)
plt.legend()
plt.show()
```

```
# @title Interpretación del parámetro beta

fig = plt.figure(figsize=(15,8))
for color, beta in zip(['r','b','k','g'],[0.01, 0.005, 0.004, 0.002]):
    t_array = np.arange(0,50,1)
    i_array = i(i0, beta, t_array)
    s_array = s(i0, beta, t_array)
    plt.plot(t_array, i_array, c = color, label = 'infectados (beta={})'.format(beta))
plt.grid(True)
plt.legend(fontsize=25)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.show()
```
